##3. Find the team has had the lowest performance (lowest volume) since Q2.2023. Find the category that contributes the least to that team.

```python
# Lọc dữ liệu từ Q2.2023 trở đi
filtered_data = payment_enriched[payment_enriched['report_month'] >= '2023-04']
print(filtered_data)
```

```python
# Tính tổng volume theo đội (team_own)
team_performance = filtered_data.groupby('team_own')['volume'].sum()

# Sắp xếp theo volume tăng dần
team_performance_sort = team_performance.sort_values(ascending=True)
lowest_perform_team = team_performance_sort.idxmin()  # Lấy tên đội
lowest_perform_volume = team_performance_sort.iloc[0]  # Lấy volume của đội


print(f"Team with the lowest performance: {lowest_perform_team} with volume: {lowest_perform_volume}")
```

```python
# Lọc dữ liệu bảng filtered_data chỉ bao gồm đội có hiệu suất thấp nhất
team_data = filtered_data[filtered_data['team_own'] == lowest_perform_team]

# Tính tổng volume theo category
category_performance = team_data.groupby('category')['volume'].sum()

# Tìm category có hiệu suất thấp nhất
lowest_contrib_category = category_performance.idxmin()
print(f"Category with lowest contribution to the team: {lowest_contrib_category}")
```

##5. Define type of transactions (‘transaction_type’) for each row, given:
- transType = 2 & merchant_id = 1205: Bank Transfer Transaction
- transType = 2 & merchant_id = 2260: Withdraw Money Transaction
- transType = 2 & merchant_id = 2270: Top Up Money Transaction
- transType = 2 & others merchant_id: Payment Transaction
- transType = 8, merchant_id = 2250: Transfer Money Transaction
- transType = 8 & others merchant_id: Split Bill Transaction
- Remained cases are invalid transactions


```python
# prompt: Define type of transactions (‘transaction_type’) for each row, given:
# - transType = 2 & merchant_id = 1205: Bank Transfer Transaction
# - transType = 2 & merchant_id = 2260: Withdraw Money Transaction
# - transType = 2 & merchant_id = 2270: Top Up Money Transaction
# - transType = 2 & others merchant_id: Payment Transaction
# - transType = 8, merchant_id = 2250: Transfer Money Transaction
# - transType = 8 & others merchant_id: Split Bill Transaction
# - Remained cases are invalid transactions

def categorize_transaction(row):
    if row['transType'] == 2:
        if row['merchant_id'] == 1205:
            return 'Bank Transfer Transaction'
        elif row['merchant_id'] == 2260:
            return 'Withdraw Money Transaction'
        elif row['merchant_id'] == 2270:
            return 'Top Up Money Transaction'
        else:
            return 'Payment Transaction'
    elif row['transType'] == 8:
        if row['merchant_id'] == 2250:
            return 'Transfer Money Transaction'
        else:
            return 'Split Bill Transaction'
    else:
        return 'Invalid Transaction'

transactions['transaction_type'] = transactions.apply(categorize_transaction, axis=1)
transactions
```

##6. Of each transaction type (excluding invalid transactions): find the number of transactions, volume, senders and receivers.


```python
# Lọc các giao dịch không phải Invalid Transaction
valid_transactions = transactions[transactions['transaction_type'] != 'Invalid Transaction']

# Nhóm theo transaction_type và tính toán các giá trị cần thiết
transaction_summary = valid_transactions.groupby('transaction_type').agg(
    num_transactions=('transaction_type', 'count'),  # Số lượng giao dịch
    total_volume=('volume', 'sum'),  # Tổng volume
    num_senders=('sender_id', 'nunique'),  # Số lượng senders duy nhất
    num_receivers=('receiver_id', 'nunique')  # Số lượng receivers duy nhất
)

print(transaction_summary)
```
